```python


plt.title('box plot of the data to show the variation b/w distance and fair of taxi')
plt.xlabel('total_distance')
plt.ylabel('taxi_fair')


yellowtaxi.boxplot()


plt.show()
```

```python
yellowtaxi.plot(kind='box', figsize=(12,8))
plt.show()
```

**Box Plots** :: *Boxplots are a measure of how well distributed the data in a data set is. It divides the data set into three quartiles. This graph represents the minimum, maximum, median, first quartile and third quartile in the data set.*

```python
plt.rcParams['agg.path.chunksize'] = 1000
plt.title('2d plot of the data to show the variation b/w distance and fair of taxi')
plt.xlabel('total_distance')
plt.ylabel('taxi_fair')
plt.plot(yellowtaxi['total_distance'], yellowtaxi['taxi_fair'], color='blue')
plt.show()
```

**this 2d plot is showing the variation of total distance travelled and the taxi fair**
